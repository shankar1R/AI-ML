```python
import pandas as pd

bmi={ "age":[20,23,21,24,23],"gender":["M","F","M","F","M"],"height":[161,164,166,170,169],"weight":[46,66,68,71,74] }

df=pd.DataFrame(bmi)
# print(df)
def healthstatus(bmi):
    status=""
    if (bmi < 18.5):
        print("underweight")
        status="UnderWeight"
        return status
  
    elif (bmi >= 18.5 and bmi < 24.9):
        print("Healthy")
        status="Healthy"
        return status
  
  
    elif (bmi >= 24.9 and bmi < 30):
        print( "overweight")
        status="overweight"
        return status
  
    elif (bmi >= 30):
        print("Suffering from Obesity")
        status="Suffering from Obesity"
        return status
  
bmivalues=[]
healthstatuses=[]
for i in range(1):
    ht=bmi["height"]
    wt=bmi["weight"]
    
    for j in range(len(ht)):
        # print(ht[j])
        # print(wt[j])
        ans=wt[j]/(ht[j]/100)**2
        print(ans)
        bmivalues.append(ans)
        bmistate=healthstatus(ans)
        healthstatuses.append(bmistate)

df["BMI Values"]=bmivalues
df["Your Health Status"]=healthstatuses
df
```
